```python
import yfinance as yf
from talib.abstract import EMA,DEMA,TEMA
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2020-05-10", end="2022-09-10")
```

```python
ema100 = EMA(data['Close'],100)
ema200 = EMA(data['Close'],200)
```

```python
ema100[-2]
```

```python
ema100[-1]
```

```python
fig = plt.figure(figsize=(10, 7))
plt.title('stochastic')
plt.xlabel('Date')
plt.ylabel('MFI values')

plt.plot(ema100[-200:], 'm', label='ema100')
plt.plot(ema200[-200:], 'r', label='ema200')
```

```python
dema100 = DEMA(data['Close'],100)
dema200 = DEMA(data['Close'],200)
```

```python
fig = plt.figure(figsize=(10, 7))
plt.title('stochastic')
plt.xlabel('Date')
plt.ylabel('MFI values')

plt.plot(dema100, 'm', label='dema100')
plt.plot(dema200, 'r', label='dema200')
```

```python
tema100 = TEMA(data['Close'],100)
tema200 = TEMA(data['Close'],200)
```

```python
fig = plt.figure(figsize=(10, 7))
plt.title('stochastic')
plt.xlabel('Date')
plt.ylabel('MFI values')

plt.plot(tema100, 'm', label='tema100')
plt.plot(tema200, 'r', label='tema200')
```
