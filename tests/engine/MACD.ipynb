```python
import yfinance as yf
from talib.abstract import MACD, MACDEXT, MACDFIX
import matplotlib.pyplot as plt
```

```python
data = yf.download("AAPL", start="2020-05-10", end="2022-09-10")
```

```python
macd, macdsignal, _ = MACD(data['Close'])
```

```python
fig = plt.figure(figsize=(10, 7))
plt.title('MACD')

plt.plot(macd, 'm', label='macd')
plt.plot(macdsignal, 'r', label='macdsignal')
```

```python
macd, macdsignal, _ = MACDEXT(data['Close'])
```

```python
fig = plt.figure(figsize=(10, 7))
plt.title('MACDEXT')

plt.plot(macd, 'm', label='macd')
plt.plot(macdsignal, 'r', label='macdsignal')
```

```python
macd, macdsignal, _ = MACDFIX(data['Close'])
fig = plt.figure(figsize=(10, 7))
plt.title('MACDFIX')

plt.plot(macd, 'm', label='macd')
plt.plot(macdsignal, 'r', label='macdsignal')
```
